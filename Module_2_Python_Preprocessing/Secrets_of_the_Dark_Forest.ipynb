```python
# Обработка 'name'
c_df = c_df.dropna(subset=['name'])
```

```python
# Обработка 'year_of_release'
# Для игр где year_of_release = NaN, но в названии есть приписка о годе (например 'Madden NFL 2004'), подстваляем этот год в year_of_release

def check_year(row):
    try:
        year = int(row['name'][-4:])
        # Проверяем, что это год
        if year >= 1960 and year <= 2025:
            return year
        else:
            return row['year_of_release']
    except:
        return row['year_of_release']

c_df['year_of_release'] = c_df.apply(check_year, axis=1)

# Заполнение средним значением года по платформе для всех оставшихся year_of_release = NaN
def year_by_platform(row):
    if pd.isna(row['year_of_release']):
        group = c_df[c_df['platform'] == row['platform']]
        return round(group['year_of_release'].mean(), 0)
    else:
        return row['year_of_release']
    
c_df['year_of_release'] = c_df.apply(year_by_platform, axis=1)

# Меняем тип данных с float на int
c_df['year_of_release'] = c_df['year_of_release'].astype('int64')
```

```python
# Обработка eu_sales

def eu_sales_filler(row):
    if pd.isna(row['eu_sales']):
        group = c_df[
            (c_df['platform'] == row['platform']) &
            (c_df['year_of_release'] == row['year_of_release']) &
            (c_df['genre'] == row['genre'])
        ]
        
        mean_sales = group['eu_sales'].mean()
        return mean_sales
    else:
        return row['eu_sales']
    
c_df['eu_sales'] = c_df.apply(eu_sales_filler, axis=1)
```

```python
# Обработка jp_sales

def jp_sales_filler(row):
    if pd.isna(row['jp_sales']):
        group = c_df[
            (c_df['platform'] == row['platform']) &
            (c_df['year_of_release'] == row['year_of_release']) &
            (c_df['genre'] == row['genre'])
        ]
        
        mean_sales = group['jp_sales'].mean()
        return mean_sales
    else:
        return row['jp_sales']

c_df['jp_sales'] = c_df.apply(jp_sales_filler, axis=1)
```

```python
# Обработка critic_score
c_df['critic_score'] = c_df['critic_score'].fillna(-1)
```

```python
# Обработка user_score
c_df['user_score'] = c_df['user_score'].fillna(-1)
```

```python
# Обработка rating

# Прежде чем применять алгоритм проврерим значения
print(c_df['rating'].unique()) # Найдено устаревшее значение 'K-A', в последствии его заменили на 'E', сделаем также
print()
print(c_df['genre'].unique()) # Видно, что значения повторяются в разных регистрах, нужно привести все к одному
```

```python
# Замена устаревшего рейтинга 'K-A' на 'E'
def rating_replace(row):
    if row['rating'] == 'K-A':
        return 'E'
    else:
        return row['rating']
    
c_df['rating'] = c_df.apply(rating_replace, axis=1)
    
    
# Привеодим 'genre' к нижнему регистру
c_df['genre'] = c_df['genre'].str.lower()


# Определние текущего распределения рейтинга для каждого жанра
rating_prob_by_genre = {}

for genre in c_df['genre'].unique():
    rating_prob_by_genre[genre] = c_df[c_df['genre']==genre].groupby(by='rating')['genre'].count()
    / c_df[c_df['genre']==genre].groupby(by='rating')['genre'].count().sum()

    
# Принимаем  распределние рейтинга по жанрам за вероятность того, что оно окажется в пустых ячейках и заполняем
import random

def get_rating(row):
    if pd.isna(row['rating']):
        cur_genre = row['genre']
        genre = rating_prob_by_genre[cur_genre].keys().to_list()
        weights = rating_prob_by_genre.get(cur_genre).to_list()
        res = random.choices(genre, weights=weights, k=1)
        return res[0]

    else:
        return row['rating']

c_df['rating'] = c_df.apply(get_rating, axis=1)
```
