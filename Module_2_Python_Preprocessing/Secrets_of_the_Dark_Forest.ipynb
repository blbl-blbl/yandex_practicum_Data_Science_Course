Кроме NaN в столбце user_score также есть значения 'tbd', значит для некоторых игр user_score еще уточняется. В данном случае, зная, что в данном столбце только положительные значения, заменить пустые строки и строки с 'tbd' на - 1.

```python
# Замена NaN и 'tbd' на -1

c_df['user_score'] = c_df['user_score'].fillna(-1)

def tbd_replace(row):
    if row['user_score']=='tbd':
        return -1
    else:
        return row['user_score']

c_df['user_score'] = c_df.apply(tbd_replace, axis=1)

# Преобразуем в float и заодно проверяем что нет никаких проблем со значениями
c_df['user_score'] = c_df['user_score'].astype('float64')
```

#### Проверка столбца rating

```python
c_df['rating'].unique()
```

Уже знаем, что в этом столбце есть пустые строки, теперь еще можем заметить, что существуют строки с устаревшим рейтингом 'K-A'. В 1998 году этот рейтинг был заменен на 'E'. Значит в датафрейме нужно изменить устаревший рейтинг на новый.

```python
print(f"Количество строк с рейтингом 'K-A': {len(c_df[c_df['rating']=='K-A'])}")
c_df[c_df['rating']=='K-A']
```

```python
# Замена устаревшего рейтинга 'K-A' на 'E'
def rating_replace(row):
    if row['rating'] == 'K-A':
        return 'E'
    else:
        return row['rating']

c_df['rating'] = c_df.apply(rating_replace, axis=1)
```

```python
print(f"Количество строк с пустыми значениями в 'rating': {len(c_df[c_df['rating'].isna() == True])}\n")
c_df[c_df['rating'].isna() == True].head()
```

Для того чтобы заменить пустые строки в rating применяем следующий алгоритм - для каждого жанра находим долю каждого рейтинга и принимаем, что в пустых строках сохранится такое же распределение т.е. в rating_prob_by_genre находятся вероятности рейтинга для каждого жанра

```python
rating_prob_by_genre = {}
for genre in c_df['genre'].unique():
    rating_prob_by_genre[genre] = c_df[c_df['genre']==genre].groupby(by='rating')['genre'].count() / c_df[c_df['genre']==genre].groupby(by='rating')['genre'].count().sum()

import random

def get_rating(row):
    if pd.isna(row['rating']):
        cur_genre = row['genre']
        genre = rating_prob_by_genre[cur_genre].keys().to_list()
        weights = rating_prob_by_genre.get(cur_genre).to_list()
        res = random.choices(genre, weights=weights, k=1)
        return res[0]

    else:
        return row['rating']

c_df['rating'] = c_df.apply(get_rating, axis=1)
```
